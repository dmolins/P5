```python
from matplotlib import pyplot as plt 
import numpy as np 
```

```python
def openText(filename):
    text = open(filename, 'r')
    text = text.readlines()
    values = []
    for l in text:
        values.append(float(l))
    return np.array(values)
```

```python
x = openText("x.txt")
tbl = openText("table.txt")
plt.plot(x, 'ro', markersize=3)
plt.plot(tbl, 'k-')
plt.legend(["Señal generada", "Tabla"])
plt.autoscale(enable=True, axis='x', tight=True)
plt.show()
```
